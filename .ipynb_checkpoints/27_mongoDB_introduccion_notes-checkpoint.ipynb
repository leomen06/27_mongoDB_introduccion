### <a name="mark_01"></a>Tipos de Bases NoSQL.
### [Index](#index)


- Documentales: 
    - Se empareja cada clave con una estructura de datos compleja que se denomina documento, las cuales son guardadas en **colecciones**, existen otras bases de datos documentales como Cloud Firebase o Couchbase.

- Grafos: 
    - Se utilizan para almacenar informaci√≥n sobre redes de datos, como las conexiones sociales. Neo4j es uno de los ejemplos mas populares de bases datos de este tipo.
    
- Clave - Valor: 
    - Son las bases de datos NoSQL mas simples. Cada elemento de la base de datos se almacena como un nombre de atributo (o ‚Äúclave‚Äù), junto con su valor para acceso de memoria r√°pida. un ejemplo de estas bases de datos es redis.
    
- Orientada a columnas: 
    - Estas bases de datos, como Cassandra (la mas popular) o Hbase, permiten realizar consultas en grandes conjuntos y almacenan los datos en columnas, en lugar de filas.
    
### Las bases de datos NoSQL tienen la caracter√≠stica de tener gran facilidad de escalamiento.

- Escalamiento vertical: 
    - Incrementar las caracter√≠sticas de esa misma maquina (+RAM, +CPU, +almacenamiento)

- Escalamiento horizontal: 
    - Tener una maquina (o nodo) y ‚Äúcopiarla‚Äù (ahora tienes varios nodos o maquinas con las mismas caracter√≠sticas), asegurando alta disponibilidad, sistemas de replicacion o tener un conjunto que responda en simultaneo. Esto te libera de la necesidad de escalar de forma vertical.

### Hacer escalamiento vertical es mas f√°cil pero mas costoso con el paso del tiempo. el escalamiento horizontal es mas costoso en el inicio pero luego el precio se mantiene ‚Äúest√°ndar‚Äù con el paso del tiempo.

![](img_01.png)
**---------------------------------------------------------------------------------------------------------------------------------------------------------**

- Las bases de datos NoSQL aprovechan muy bien el Escalamiento horizontal: puedes armar un cluster de informaci√≥n en el que los datos se van a conectar entre si (varios nodos). Esto hace que los bases datos NoSQL sea mas f√°cilmente escalable o replicable que una SQL, debido a que hablamos de documentos y no bases de datos relacionadas.

- Que es Replicacion? 
    - Es una t√©cnica en la que, una ves distribuidas la base de datos en varios nodos (EHüòâ), mediante un **load balancer**, asigna las peticiones y consultas a cada uno de los nodos en una forma ordenada. Esto permite procesamiento en paralelo y tambi√©n alta disponibilidad (si una replica falla, habr√° otra disponible). Esto no es posible en el escalamiento vertical.

### <a name="mark_09"></a>JSON vs. BSON.
### [Index](#index)

Documentaci√≥n https://www.mongodb.com/basics/bson

- JSON

Ventajas de JSON

1. Amigable.
2. Se puede leer.
3. Es un formato muy usado.

- Desventajas de JSON

1. Basado en texto.
2. Consume mucho espacio.
3. Es limitado: string, boolean, number, arrays.

- BSON

Ventajas de BSON

1. Representaci√≥n binaria de JSON.
2. Consume menos espacio.
3. Alto rendimiento.
4. Tipos de datos: date, raw binary, integer, long, float, y m√°s.

Desventajas de BSON

1. No es est√°ndar.
2. Es un formato hecho para la m√°quina, no para leer.

### <a name="mark_12"></a>updateOne(), "set", "inc" actualizando un documento.
### [Index](#index)

- En la actualizaci√≥n, se puede cambiar el valor de una llave, o agregar una nueva llave:valor.
- Primero indico el id del documento a actualizar
- Segundo aplico los cambios con $set.


```json
use("platzi_store")

db.products.updateOne(
  // query
  {_id: 2},
  // change => operators
  {
    $set: {
      name: 'Change name',
      price: 3000,
      tags: ['A', 'B', 'C']
    }
  }
)

```

### El otro operador $inc, es para n√∫meros e incrementa un valor.

```json
use("platzi_store")

db.products.updateOne(
  // query
  {_id: 2},
  // change => operators
  {
    $inc: {
      price: 100,
    }
  }
)

```
- Incrementa el precio en 100.

### Para hacer un update por ObjectId.

```json
use("platzi_store")

db.products.updateOne(
  // query
  {_id: ObjectId("62f573dcab088f7ebb862f75")},
  // change => operators
  {
    $inc: {
      price: 100,
    }
  }
)

```

### <a name="mark_14"></a>Listado de operadores de actualizaci√≥n:
### [Index](#index) 

1. ```$inc```: Incrementa el valor de un atributo num√©rico en una cantidad espec√≠fica. 

2. ```$mul```: Multiplica el valor de un atributo num√©rico por un factor espec√≠fico.

3. ```$rename```: Cambia el nombre de un atributo. 

4. ```$set```: Asigna un valor espec√≠fico a un atributo.

5. ```$unset```: Elimina un atributo de un documento.

6. ```$min```: Actualiza el valor de un atributo con el valor m√≠nimo especificado, s√≥lo si el valor actual es mayor que el valor especificado.

7. ```$max```: Actualiza el valor de un atributo con el valor m√°ximo especificado, s√≥lo si el valor actual es menor que el valor especificado.

8. ```$currentDate```: Establece el valor de un atributo como la fecha y hora actual.

9. ```$addToSet```: A√±ade un valor a un atributo de tipo conjunto (array), s√≥lo si el valor no existe en el conjunto.

10. ```$pop```: Elimina el primer o √∫ltimo elemento de un atributo de tipo conjunto (array).

```json
db.mi_colleccion.updateOne({
   sensor: "A001",
   date: "2022-01-03",
 }, {
   $pop: {
     readings: 1
   }
 })
```

- Nota: para este caso el documento tiene 3 valores "sensor", "date" y "reading" que es una array, con "readings: 1" elimina el √∫ltimo valor del array y con "reading:-1" elimina el primero.


12. ```$pull```: Elimina un valor espec√≠fico de un atributo de tipo conjunto (array).

- Quitando el elemento "book" de cualquier documento "{}", para el array "tag" que lo pueda contener.

```json
use("platzi_store")

db.inventory.updateMany({}, {
  $pull: {
    tags: "book"
  }
})
```

- Quitando varios elementos "book", y "school" de cualquier documento "{}", para el array "tag" que los contenga.

```json
use("platzi_store")

db.inventory.updateMany({},{
    $pull:{
        tags:{
            $in: ["book", "school"]
        }
    }
})
```

13. ```$push```: A√±ade un valor (al final) a un atributo de tipo conjunto (array).

- Inserta un nuevo valor "headphone" al array existente "tags".

```json
use("platzi_store")

db.inventory.updateOne({_id: 4}, {
  $push: {
    tags: "headphone"
  }
})
```

14. ```$pullAll```: Elimina varios valores espec√≠ficos de un atributo de tipo conjunto (array).

### <a name="mark_15"></a>Update or insert:
### [Index](#index) 

- Para un data set que se ve de la siguiente forma:

```json
use("platzi_store")

db.iot.drop()

db.iot.insertMany([
  { _id: 1, sensor: "A001", date: "2022-01-01", readings: [1,2,3,4] },
  { _id: 2, sensor: "A001", date: "2022-01-02", readings: [1,2,3,4] },
  { _id: 3, sensor: "A002", date: "2022-01-01", readings: [1,2,3,4] },
  { _id: 4, sensor: "A002", date: "2022-01-02", readings: [1,2,3,4] },
])

db.iot.find()
```

- En este caso particular estamos buscando.
    - Qu√© se creo o inserte un nuevo dato en el array de "reading" si este documento NO Existe.
    - Que se actualize el array de "reading" si el document SI Existe.
- Entonces la combinaci√≥n de "updateOne()", el operador #push, con el agregado del par√°metro de "upsert: true", se encarga de realizar un insert o un update automaticamente dependiendo el caso.

```json
use("platzi_store")

db.iot.updateOne({
  sensor: "A001",
  date: "2022-01-04",
}, {
  $push: {
    readings: 12434
  }
}, {
  upsert: true
})

```

### <a name="mark_17"></a> Operadores de comparaci√≥n `$eq` (equal) y `$ne` (non equal)
### [Index](#index) 



- Supongamos tener el siguente data set.
```json
use("platzi_store")

db.inventory.drop()

db.inventory.insertMany([
  { _id: 1, item: { name: "ab", code: "123" }, qty: 15, tags: ["A", "B", "C"] },
  { _id: 2, item: { name: "cd", code: "123" }, qty: 20, tags: ["B"] },
  { _id: 3, item: { name: "ij", code: "456" }, qty: 25, tags: ["A", "B"] },
  { _id: 4, item: { name: "xy", code: "456" }, qty: 30, tags: ["B", "A"] },
  { _id: 5, item: { name: "mn", code: "000" }, qty: 20, tags: [["A", "B"], "C"] },
])

db.inventory.find()
```

- Primer ejemplo, utilizando `$eq` busco "qty:20", notar que las sentencias comentadas dan el mimso resultado, porque el operador `$eq` ya est√° implicito en mongo.

```json
use("platzi_store")

//db.inventory.find({ qty: 20 })
db.inventory.find({ qty: { $eq: 20 } })
```

- Segundo ejemplo:

```json
use("platzi_store")

// db.inventory.find({ "item.code": "123" })
db.inventory.find({ "item.code": { $eq: "123" } })
```

- Ejemplo con `$ne`:
```json
use("platzi_store)

db.inventory.find(qty: {$ne: 20})
```

- Ejemplo convinando con updateMany()

```json
use("platzi_store")

db.inventory.updateMany(
  // query - operators
  { qty: { $ne: 20 } },
  // update - operators
  {
    $inc: {
      qty: 10
    }
  }
)
```

### <a name="mark_18"></a> Operadores de comparaci√≥n `$gt` (greater than), `$gte` (greater than or equal), `$lt` less than, `$lte` less than or equal.

- $lte (less than or equal - <=)


### <a name="mark_19"></a> Combinando operadores `$gt` (greater than), `$gte` (greater than or equal), `$lt` less than, `$lte` less than or equal.
### [Index](#index) 

- Combinaci√≥n simple.
```json
use(platzti_store)

db.inventory.find(qty: {$gt:20, $lt:80})
```

- Combinaci√≥n un poco m√°s compleja.
```json
use(platzi_store)

db.inventory.find({
  "item.code": { $ne: "123" },
  qty: { $gte: 20, $lte: 25 }
})
```

- Combinado con un updateMany() quitando con $pull los valores mayores o iguales a 3 en el array "readings"

```json
use("platzi_store")

db.iot.updateMany(
  // query
  { sensor: "A001" },
  // update
  {
    $pull: {
      readings: { $gte: 3 }
    }
  }
)
```


### <a name="mark_20"></a> Consulta con `$regex`.
### [Index](#index) 

- Data set a ser utilizado:
```json
use("platzi_store")

db.inventory.drop()

db.inventory.insertMany([
  { _id: 1, item: { name: "ab", code: "123", description : "Single line description."    }, qty: 15, tags: ["A", "B", "C"] },
  { _id: 2, item: { name: "cd", code: "123", description : "First line\nSecond line"     }, qty: 20, tags: ["B"] },
  { _id: 3, item: { name: "ij", code: "456", description : "Many spaces before     line" }, qty: 25, tags: ["A", "B"] },
  { _id: 4, item: { name: "xy", code: "456", description : "Multiple\nline description"  }, qty: 30, tags: ["B", "A"] },
  { _id: 5, item: { name: "mn", code: "000" }, qty: 20, tags: [["A", "B"], "C"] },
])

db.inventory.find()
```
- Consulta sobrela propiedad "desciption", buscando los que tienen la palabra "line".

```json
use("platzi_store")

db.inventory.find({"item.description": {$regex: /line/}})
```

- Si no queremos que sea sensitive case colocamos un "i" al final de /LINE/i
```json
use("platzi_store")

db.inventory.find({"item.description": {$regex: /LINE/i}})
```

- Si buscamos los documentos en los cuales la oraci√≥n termina con la palabra "line" --> /line$/i
```json
use("platzi_store")

db.inventory.find({"item.description": {$regex: /line$/i}})
```

- Si buscamos los documentos en los cuales la oraci√≥n inicie con la palabra "single" --> /^single/i
```json
use("platzi_store")

db.inventory.find({"item.description": {$regex: /^single/i}})
```

- Si la oraci√≥n/ contiene un salta de line "\n" para el caso "First line\nSecond line", la segunda parte de la oraci√≥n comiensa con "s", entonces si buscamos todas las palabras que inicien con "s" teniendo en cuenta el salto de l√≠nea usamos --> /^s/im.

```json
use("platzi_store")

db.inventory.find({"item.description": {$regex: /^s/im}})
```

### <a name="mark_20"></a> Projection, seleccionando que mostrar
### [Index](#index) 


- En el siguente ejemplo, se seleccionan solo las propiedades "tripduration" y "usertype", que ser√°n mostradas luego de realizada la consulta.

```json
use("sample_training")

db.trips.find(
  // query
  { tripduration: { $gte: 500 } },
  // projection
  { tripduration: 1, usertype: 1 }
)
```

- Nota: el _id viene siempre implicito, si queremos eliminarlo de los resultado de la consulta, debemos hacerlo explicitamente.

```json
use("sample_training")

db.trips.find(
  // query
  { tripduration: { $gte: 500 } },
  // projection
  { tripduration: 1, usertype: 1, _id: 0 }
)
```

### <a name="mark_21"></a> Operadores para arrays.


### <a name="mark_22"></a> Operadores l√≥gico `$and`, `$or`, `$nor`, y `$not`.
### [Index](#index) 

- $and => que cumpla las dos o m√°s condiciones dadas.

```json
use("sample_training")

// and implict
 db.inspections.find({
   sector: "Tax Preparers - 891",
   result: "Unable to Locate"
 }).count() // 6

// and explict
// $and [{}, {}, {}]
 db.inspections.find({
   $and: [
     { sector: "Tax Preparers - 891" },
     { result: "Unable to Locate" }
   ]
 }).count() // 6
```

- $or => los que cumplan al menos una condici√≥n.

```json
use("sample_training")

// or
// $or [{}, {}, {}]
db.inspections.find({
    $or: [
         { sector: "Tax Preparers - 891" },
         { result: "Unable to Locate" }
       ]
     }).count()
```

- $nor => los que no cumplan con ninguna condici√≥n.

```json
use("sample_training")

// nor
// $nor [{}, {}, {}]
 db.inspections.find({
   $nor: [
     { sector: "Tax Preparers - 891" },
     { result: "Unable to Locate" }
   ]
 })
 db.inspections.find({
   $nor: [
     { result: "No Violation Issued" },
     { result: "Unable to Locate" }
   ]
 } , {
   result: 1,
   _id: 0
 })
```

- $not => excluye aplica valores en especifico y no tiene forma de array

```json
use("sample_training")

// $not = {}
db.inspections.find({
  result: { $not: { $regex: /Unable to Locate/ } }
})

```

- El potencial de esta funcionalidad radica en poder convertir a las propiedades de un documento en operadores, mediante la utilizaci√≥n de `$` y de esa forma podemos realizar comparaciones, no solo contra valores est√°ticos, sino entre valores de un documento de forma d√≠namica.

```json
use("platzi_store")

db.monthlyBudget.find({
  $expr: {
    $gte: ["$spent", "$budget"]// spent >= budget
  }
})

```

- Otro ejemplo utilizando un data set, pero se puede observar la convinaci√≥n de operadores.

```json
use("sample_training")

// ej_01
 db.trips.find({
   $expr: {
     $eq: ["$start station id", "$end station id"]
   }
 })
 .count()

//ej_02
db.trips.find({
  $expr: {
    $and: [
      { $eq: ["$start station id", "$end station id"] },
      { $gte: ["$tripduration", 1200] }
    ]
  }
})
.count()

```

### <a name="mark_26"></a> Sort, limit y skip.
### [Index](#index) 

- .sort() => para ordenar los doc, con un 1 de menor a mayor o de la A a la Z y con un -1 al contrario ( 10-0, Z-A).

- .limit() => Limita la cantidad de doc que queremos traer, ej: .limit(3) trae 3 docs.

- En el siguiente ejemplo, vemos como primero se realiza un find la condici√≥n `$gte: 100`, luego se ordena de manera descendente, y finalmente se muestran solamente los 2 primeros documentos.

```json
use("sample_training")

// 1 = 0 - 10, A-Z
// -1 = 10 - 0, Z-A
db.zips
  .find({ pop: { $gte: 100 } })
  .sort({ pop: -1 })
  .limit(2)

```
- .skip() => indica desde que posici√≥n nos traera los doc, ej: en una lista del 1 al 10; .skip(3) nos da los doc desde el 4 hasta el 10.

- En el siguiente ejemplo, se buscan todos los documentos, se limpian los primeros 9 y finalmente de todos esos documentos se muestran solo 3, los cuales ser√≠an documentos 10, 11 y 12. Esto tambi√©n se puede denominar "paginaci√≥n"

```json
use("platzi_store")

db.categories
  .find()
  .skip(9)
  .limit(3)

```
